# Comparing XAI Methods

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import torch
from cnn2 import AlexNet

import warnings
warnings.filterwarnings("ignore")
```

```python
data_path = "../../../data/"
models_path = data_path + "xai/nih/models/"
models_folder = os.listdir(models_path)
model_list = []
methods = ["ig","gs","occ","nt"]
for mo in models_folder:
    print(mo)
    model = AlexNet().to("cpu")
    model.load_state_dict(torch.load(models_path + f"{mo}", map_location=torch.device('cpu')))
    model = model.eval()
    model_list.append(model)
    del model
```

```python
def plot_dist(df,c,h):
    #df = pd.DataFrame(df)
    mx = df.max().max()
    mn = df.min().min()
    fig = plt.figure(figsize=(20,4))
    fig.subplots_adjust(hspace=0.4, wspace=0.4)
    for i in range(1,len(methods)+1):
        ax = fig.add_subplot(1, len(methods)+1, i)
        sns.distplot(df[methods[i-1]],ax=ax,bins=15,color=c)
        ax.set(ylim=(0, h),xlim=(mn,mx))
    plt.show()
```

```python
df_ig = pd.read_csv("results/ig/allresults_entropy.csv")
df_ig = df_ig.drop(columns=["Unnamed: 0"])

df_gs = pd.read_csv("results/gs/allresults_entropy.csv")
df_gs = df_gs.drop(columns=["Unnamed: 0"])

df_occ = pd.read_csv("results/occ/allresults_entropy.csv")
df_occ = df_occ.drop(columns=["Unnamed: 0"])

df_nt = pd.read_csv("results/nt/allresults_entropy.csv")
df_nt = df_nt.drop(columns=["Unnamed: 0"])
```

```python
df_m1 = pd.DataFrame()
df_m1["ig"] = df_ig["model1"]
df_m1["gs"] = df_gs["model1"]
df_m1["occ"] = df_occ["model1"]
df_m1["nt"] = df_nt["model1"]
```

```python
plot_dist(df_m1,"steelblue",2)
```

```python
fig, ax = plt.subplots()
ax.set(ylim=(0, 6))
sns.violinplot(data=df_m1)
```
